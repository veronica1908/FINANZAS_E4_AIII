## Selección de variables

```python
# Ahora, utiliza estos modelos para seleccionar las variables
var_names = funciones.sel_variables(modelos_reg, X, y, threshold="2*mean")
X2=X[var_names] ### matriz con variables seleccionadas
X2.info()
X.info()
```

## Definir modelo ganador

```python
# Dividir los datos en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X2, y, test_size=0.2, random_state=42)

# Inicializar los modelos
models = {
    "Decision Tree": DecisionTreeRegressor(),
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42)
}

# Entrenar y evaluar cada modelo
results = {}

for name, model in models.items():
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    results[name] = {"MSE": mse, "R2": r2}

# Mostrar los resultados
for name, metrics in results.items():
    print(f"{name}: MSE = {metrics['MSE']:.4f}, R2 = {metrics['R2']:.4f}")
```

Dado que los resultados de los modelos son muy similares y no muestran un rendimiento significativo (con valores de R² muy bajos y MSE altos), parece que ninguno de los modelos está funcionando adecuadamente con los datos actuales. Vamos a realizar algunos ajustes a ontinuación.

```python
# Evaluamos el modelo usando validación cruzada
scores = cross_val_score(GradientBoostingRegressor(), X2, y, cv=5, scoring='neg_mean_squared_error')
print("MSE promedio:", -scores.mean())
```

El resultado de la validación cruzada con un MSE promedio de aproximadamente 7.68 billones sigue siendo bastante alto, indicando que los modelos aún tienen problemas para predecir con precisión el valor pagado. Vamos a afinar hiperparámateros a ver si se obtiene alguna mejora significativa.

##Afinar hiperparámetros

```python
# Probamos afinar el random forest con R2
# Definir el modelo
rf_model = RandomForestRegressor()

# Definir el grid de hiperparámetros con distribuciones para la búsqueda aleatoria
param_dist = {
    'n_estimators': randint(100, 300),
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': randint(10, 30),
    'min_samples_split': randint(2, 10),
    'min_samples_leaf': randint(1, 4),
    'bootstrap': [True, False]
}

# Configurar la búsqueda aleatoria
random_search = RandomizedSearchCV(estimator=rf_model, param_distributions=param_dist,
                                   n_iter=50, cv=3, scoring='r2', n_jobs=-1, random_state=42)

# Realizar la búsqueda
random_search.fit(X, y)

# Obtener los mejores hiperparámetros
best_params = random_search.best_params_
best_score = random_search.best_score_

print(f"Mejores hiperparámetros: {best_params}")
print(f"Mejor R2: {best_score}")
```

```python
#Vamos a realizar el afinamiento de hiperparámetros para para GradientBoostingRegressor
m_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [3, 4, 5],
    'min_samples_split': [2, 5, 10],
    'learning_rate': [0.01, 0.1, 0.2]
}

grid_search = GridSearchCV(estimator=GradientBoostingRegressor(), param_grid=param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(X_test)
final_mse = mean_squared_error(y_test, y_pred)
final_r2 = r2_score(y_test, y_pred)
print(f"Mejor modelo: MSE = {final_mse:.4f}, R2 = {final_r2:.4f}")
```
